```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import warnings
warnings.filterwarnings('ignore')

from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from scipy.spatial.distance import squareform

np.random.seed(42)

plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 8)

print("Librerías importadas correctamente")
print("Semilla aleatoria fijada en 42 para reproducibilidad")
```

```python
print("CARGANDO DATOS ORIGINALES Y MATRICES DE DISTANCIA")
print("="*60)

column_names = [
    'age', 'workclass', 'fnlwgt', 'education', 'education_num',
    'marital_status', 'occupation', 'relationship', 'race', 'sex',
    'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income'
]

df = pd.read_csv('../../../datos/adult.data', names=column_names, na_values=' ?')

print(f"Dataset original cargado: {df.shape[0]} filas, {df.shape[1]} columnas")

categorical_missing = ['workclass', 'occupation', 'native_country']
for col in categorical_missing:
    if df[col].isnull().sum() > 0:
        mode_value = df[col].mode()[0]
        df[col].fillna(mode_value, inplace=True)

df['income_binary'] = (df['income'] == ' >50K').astype(int)
df['sex_binary'] = (df['sex'] == ' Female').astype(int)

print("\nCargando matrices de distancia de Gower...")
distance_matrix_5k = np.load('../ej3/distance_matrix_5000.npy')
distance_matrix_10k = np.load('../ej3/distance_matrix_10000.npy')

print(f"Matriz 5K: {distance_matrix_5k.shape}")
print(f"Matriz 10K: {distance_matrix_10k.shape}")

print("\nDatos cargados correctamente")
```

```python
print("SELECCIÓN DE MUESTRAS (CONSISTENTE CON EJERCICIO 3)")
print("="*60)

sample_5k_indices = df.sample(n=5000, random_state=42).index.tolist()
sample_10k_indices = df.sample(n=10000, random_state=42).index.tolist()

sample_5k_data = df.loc[sample_5k_indices].copy()
sample_10k_data = df.loc[sample_10k_indices].copy()

print(f"Muestra 5K: {len(sample_5k_data)} registros")
print(f"Muestra 10K: {len(sample_10k_data)} registros")

print("\nÍndices de muestra 5K (primeros 10):", sample_5k_indices[:10])
print("Índices de muestra 10K (primeros 10):", sample_10k_indices[:10])

print("\nMuestras preparadas correctamente")
```

```python
print("EVALUACIÓN DEL NÚMERO ÓPTIMO DE CLUSTERS - MUESTRA 5K")
print("="*60)

k_range = range(2, 11)
silhouette_scores_5k = []
davies_bouldin_scores_5k = []

print("Evaluando métricas para diferentes valores de k...")
for k in k_range:
    print(f"  Procesando k={k}...", end=" ")
    
    clusterer = AgglomerativeClustering(
        n_clusters=k,
        metric='precomputed',
        linkage='average'
    )
    labels = clusterer.fit_predict(distance_matrix_5k)
    
    silhouette = silhouette_score(distance_matrix_5k, labels, metric='precomputed')
    davies_bouldin = davies_bouldin_score(distance_matrix_5k, labels)
    
    silhouette_scores_5k.append(silhouette)
    davies_bouldin_scores_5k.append(davies_bouldin)
    
    print(f"Silhouette: {silhouette:.4f}, Davies-Bouldin: {davies_bouldin:.4f}")

print("\nEvaluación completada para muestra 5K")
```

```python
print("EVALUACIÓN DEL NÚMERO ÓPTIMO DE CLUSTERS - MUESTRA 10K")
print("="*60)

silhouette_scores_10k = []
davies_bouldin_scores_10k = []

print("Evaluando métricas para diferentes valores de k...")
for k in k_range:
    print(f"  Procesando k={k}...", end=" ")
    
    clusterer = AgglomerativeClustering(
        n_clusters=k,
        metric='precomputed',
        linkage='average'
    )
    labels = clusterer.fit_predict(distance_matrix_10k)
    
    silhouette = silhouette_score(distance_matrix_10k, labels, metric='precomputed')
    davies_bouldin = davies_bouldin_score(distance_matrix_10k, labels)
    
    silhouette_scores_10k.append(silhouette)
    davies_bouldin_scores_10k.append(davies_bouldin)
    
    print(f"Silhouette: {silhouette:.4f}, Davies-Bouldin: {davies_bouldin:.4f}")

print("\nEvaluación completada para muestra 10K")
```

```python
print("DETERMINACIÓN DE K ÓPTIMO Y ANÁLISIS DE BALANCE")
print("="*60)

optimal_k_silhouette_5k = k_range[np.argmax(silhouette_scores_5k)]
optimal_k_silhouette_10k = k_range[np.argmax(silhouette_scores_10k)]

optimal_k_db_5k = k_range[np.argmin(davies_bouldin_scores_5k)]
optimal_k_db_10k = k_range[np.argmin(davies_bouldin_scores_10k)]

print("\nMUESTRA 5K:")
print(f"  Silhouette Score óptimo: k={optimal_k_silhouette_5k} (score: {max(silhouette_scores_5k):.4f})")
print(f"  Davies-Bouldin Score óptimo: k={optimal_k_db_5k} (score: {min(davies_bouldin_scores_5k):.4f})")

print("\nMUESTRA 10K:")
print(f"  Silhouette Score óptimo: k={optimal_k_silhouette_10k} (score: {max(silhouette_scores_10k):.4f})")
print(f"  Davies-Bouldin Score óptimo: k={optimal_k_db_10k} (score: {min(davies_bouldin_scores_10k):.4f})")

print("\n" + "="*60)
print("ANÁLISIS DE BALANCE PARA DIFERENTES VALORES DE K")
print("="*60)

for test_k in [2, 3, 4, 5]:
    print(f"\n--- k={test_k} ---")
    
    clusterer_test = AgglomerativeClustering(
        n_clusters=test_k,
        metric='precomputed',
        linkage='average'
    )
    labels_test_5k = clusterer_test.fit_predict(distance_matrix_5k)
    
    unique, counts = np.unique(labels_test_5k, return_counts=True)
    balance_ratio_5k = np.max(counts) / np.min(counts)
    
    print(f"Muestra 5K:")
    print(f"  Distribución: {dict(zip(unique, counts))}")
    print(f"  Balance ratio: {balance_ratio_5k:.2f}x")
    

    labels_test_10k = clusterer_test.fit_predict(distance_matrix_10k)
    unique, counts = np.unique(labels_test_10k, return_counts=True)
    balance_ratio_10k = np.max(counts) / np.min(counts)
    
    print(f"Muestra 10K:")
    print(f"  Distribución: {dict(zip(unique, counts))}")
    print(f"  Balance ratio: {balance_ratio_10k:.2f}x")

print("\n" + "="*60)
print("PRUEBA CON LINKAGE='COMPLETE'")
print("="*60)
print()

test_k = 4
print(f"Probando k={test_k} con linkage='complete'...")

clusterer_test_5k = AgglomerativeClustering(
    n_clusters=test_k,
    metric='precomputed',
    linkage='complete'
)
labels_test_5k = clusterer_test_5k.fit_predict(distance_matrix_5k)

unique_test, counts_test = np.unique(labels_test_5k, return_counts=True)
balance_test_5k = np.max(counts_test) / np.min(counts_test)

print(f"Muestra 5K con linkage='complete':")
print(f"  Distribución: {dict(zip(unique_test, counts_test))}")
print(f"  Balance ratio: {balance_test_5k:.2f}x")

optimal_k_5k = 4
optimal_k_10k = 4
optimal_linkage = 'complete'

print(f"\n{'='*60}")
print(f"DECISIÓN: Usar k={optimal_k_5k} priorizando balance e interpretabilidad")
print(f"{'='*60}")
print("CONTEXTO DE DESIGUALDADES:")
print("- Las métricas numéricas (Silhouette y Davies-Bouldin) favorecen k=2")
print("- PERO k=2 produce clusters extremadamente desbalanceados (>99% vs <1%)")
print("- Esto dificulta la interpretación de desigualdades demográficas")
print()
print("TRADE-OFF DECIDIDO:")
print("- k=2: Mejor métrica numérica PERO clusters ininterpretables")
print("- k=3: Mejor balance PERO aún tiene desbalance significativo")
print("- k=4: Balance aceptable + clusters interpretables ✓")
print("- k=5+: Clusters demasiado pequeños, poco poder estadístico")
print()
```

```python
print("ANÁLISIS DE COHESIÓN DE CLUSTERS")
print("="*60)
print("Análisis complementario a Davies-Bouldin: cohesión interna vs separación entre clusters")
print()

def analyze_cluster_cohesion(distance_matrix, labels):
    """Analiza cohesión interna y separación entre clusters"""
    unique_labels = np.unique(labels)
    n_clusters = len(unique_labels)
    
    intra_cluster_distances = []
    inter_cluster_distances = []
    
    for i, label_i in enumerate(unique_labels):
        cluster_i_mask = labels == label_i
        cluster_i_indices = np.where(cluster_i_mask)[0]
        
        # Distancias intra-cluster
        if len(cluster_i_indices) > 1:
            intra = []
            for idx1 in cluster_i_indices:
                for idx2 in cluster_i_indices:
                    if idx1 < idx2:
                        intra.append(distance_matrix[idx1, idx2])
            intra_cluster_distances.extend(intra)
        
        # Distancias inter-cluster
        for j, label_j in enumerate(unique_labels):
            if i < j:
                cluster_j_mask = labels == label_j
                cluster_j_indices = np.where(cluster_j_mask)[0]
                
                inter = []
                for idx1 in cluster_i_indices:
                    for idx2 in cluster_j_indices:
                        inter.append(distance_matrix[idx1, idx2])
                inter_cluster_distances.extend(inter)
    
    return {
        'intra_mean': np.mean(intra_cluster_distances),
        'intra_std': np.std(intra_cluster_distances),
        'intra_min': np.min(intra_cluster_distances),
        'intra_max': np.max(intra_cluster_distances),
        'inter_mean': np.mean(inter_cluster_distances),
        'inter_std': np.std(inter_cluster_distances),
        'inter_min': np.min(inter_cluster_distances),
        'inter_max': np.max(inter_cluster_distances),
        'cohesion_ratio': np.mean(intra_cluster_distances) / np.mean(inter_cluster_distances) if np.mean(inter_cluster_distances) > 0 else 0
    }

# Analizar muestra 5K
print("MUESTRA 5K:")
print("-" * 60)
cohesion_5k = analyze_cluster_cohesion(distance_matrix_5k, labels_5k)
print(f"Distancia intra-cluster (dentro del mismo cluster):")
print(f"  Promedio: {cohesion_5k['intra_mean']:.4f} ± {cohesion_5k['intra_std']:.4f}")
print(f"  Rango: [{cohesion_5k['intra_min']:.4f}, {cohesion_5k['intra_max']:.4f}]")
print(f"\nDistancia inter-cluster (entre clusters diferentes):")
print(f"  Promedio: {cohesion_5k['inter_mean']:.4f} ± {cohesion_5k['inter_std']:.4f}")
print(f"  Rango: [{cohesion_5k['inter_min']:.4f}, {cohesion_5k['inter_max']:.4f}]")
print(f"\nRatio de cohesión (intra/inter): {cohesion_5k['cohesion_ratio']:.4f}")
print(f"  → Cuanto MENOR el ratio, MEJOR (clusters más cohesivos y separados)")
print(f"  → Valores <0.5 indican excelente cohesión y separación")
print(f"  → Valores cercanos a 1 indican clusters poco distintivos")

# Analizar muestra 10K
print("\n\nMUESTRA 10K:")
print("-" * 60)
cohesion_10k = analyze_cluster_cohesion(distance_matrix_10k, labels_10k)
print(f"Distancia intra-cluster (dentro del mismo cluster):")
print(f"  Promedio: {cohesion_10k['intra_mean']:.4f} ± {cohesion_10k['intra_std']:.4f}")
print(f"  Rango: [{cohesion_10k['intra_min']:.4f}, {cohesion_10k['intra_max']:.4f}]")
print(f"\nDistancia inter-cluster (entre clusters diferentes):")
print(f"  Promedio: {cohesion_10k['inter_mean']:.4f} ± {cohesion_10k['inter_std']:.4f}")
print(f"  Rango: [{cohesion_10k['inter_min']:.4f}, {cohesion_10k['inter_max']:.4f}]")
print(f"\nRatio de cohesión (intra/inter): {cohesion_10k['cohesion_ratio']:.4f}")
print(f"  → Cuanto MENOR el ratio, MEJOR (clusters más cohesivos y separados)")
print(f"  → Valores <0.5 indican excelente cohesión y separación")
print(f"  → Valores cercanos a 1 indican clusters poco distintivos")
```

```python
print("ANÁLISIS DE OUTLIERS Y CLUSTERS MINORITARIOS")
print("="*60)
print("Contexto: Identificar casos extremos que afectan el agrupamiento")
print()

# Función para analizar clusters pequeños
def analyze_small_clusters(data, labels, threshold_percent=5):
    """Analiza clusters que representan menos del threshold_percent de los datos"""
    unique, counts = np.unique(labels, return_counts=True)
    total = len(labels)
    threshold_count = total * threshold_percent / 100
    
    small_clusters = []
    for cluster_id, count in zip(unique, counts):
        if count < threshold_count:
            cluster_data = data[data['cluster'] == cluster_id]
            small_clusters.append({
                'cluster_id': cluster_id,
                'count': count,
                'percentage': count/total*100,
                'data': cluster_data
            })
    
    return small_clusters

# Analizar muestra 5K
print("MUESTRA 5K:")
print("-" * 40)
small_clusters_5k = analyze_small_clusters(sample_5k_data, labels_5k, threshold_percent=10)

if small_clusters_5k:
    print(f"Se encontraron {len(small_clusters_5k)} clusters minoritarios (<10%):")
    for sc in small_clusters_5k:
        print(f"\nCluster {sc['cluster_id']}: {sc['count']} registros ({sc['percentage']:.2f}%)")
        print("Características:")
        cluster_data = sc['data']
        print(f"  Sexo: {cluster_data['sex'].value_counts().to_dict()}")
        print(f"  Raza: {cluster_data['race'].value_counts().to_dict()}")
        print(f"  Ocupación: {cluster_data['occupation'].value_counts().to_dict()}")
        print(f"  Ingresos: {cluster_data['income'].value_counts().to_dict()}")
        print(f"  Edad promedio: {cluster_data['age'].mean():.1f}")
        print(f"  Educación promedio: {cluster_data['education_num'].mean():.1f}")
else:
    print("No se encontraron clusters minoritarios (<10%)")

# Analizar muestra 10K
print("\n\nMUESTRA 10K:")
print("-" * 40)
small_clusters_10k = analyze_small_clusters(sample_10k_data, labels_10k, threshold_percent=10)

if small_clusters_10k:
    print(f"Se encontraron {len(small_clusters_10k)} clusters minoritarios (<10%):")
    for sc in small_clusters_10k:
        print(f"\nCluster {sc['cluster_id']}: {sc['count']} registros ({sc['percentage']:.2f}%)")
        print("Características:")
        cluster_data = sc['data']
        print(f"  Sexo: {cluster_data['sex'].value_counts().to_dict()}")
        print(f"  Raza: {cluster_data['race'].value_counts().to_dict()}")
        print(f"  Ocupación: {cluster_data['occupation'].value_counts().to_dict()}")
        print(f"  Ingresos: {cluster_data['income'].value_counts().to_dict()}")
        print(f"  Edad promedio: {cluster_data['age'].mean():.1f}")
        print(f"  Educación promedio: {cluster_data['education_num'].mean():.1f}")
else:
    print("No se encontraron clusters minoritarios (<10%)")
```

```python
print("ANÁLISIS PONDERADO CON FNLWGT: COMPARACIÓN PONDERADO VS NO PONDERADO")
print("="*70)
print("Contexto: Usar fnlwgt como ponderador poblacional para interpretación de desigualdades")
print()

# Función para calcular estadísticas ponderadas
def weighted_stats(data, group_col, value_col, weight_col='fnlwgt'):
    """Calcula estadísticas ponderadas por grupo"""
    weighted_sum = (data[value_col] * data[weight_col]).sum()
    total_weight = data[weight_col].sum()
    return weighted_sum / total_weight if total_weight > 0 else 0

# Función para calcular comparación completa
def compare_weighted_unweighted(cluster_data):
    """Compara estadísticas ponderadas vs no ponderadas"""
    results = {}
    
    # Ingresos
    results['income_weighted'] = weighted_stats(cluster_data, 'cluster', 'income_binary')
    results['income_unweighted'] = cluster_data['income_binary'].mean()
    
    # Sexo
    male_weight = cluster_data[cluster_data['sex'] == ' Male']['fnlwgt'].sum()
    female_weight = cluster_data[cluster_data['sex'] == ' Female']['fnlwgt'].sum()
    total_weight = cluster_data['fnlwgt'].sum()
    
    results['male_weighted'] = male_weight/total_weight*100
    results['female_weighted'] = female_weight/total_weight*100
    results['male_unweighted'] = (cluster_data['sex'] == ' Male').mean() * 100
    results['female_unweighted'] = (cluster_data['sex'] == ' Female').mean() * 100
    
    # Ocupación top 3
    occ_unweighted = cluster_data['occupation'].value_counts(normalize=True).head(3)
    occ_weighted = pd.DataFrame({
        'occupation': cluster_data['occupation'],
        'fnlwgt': cluster_data['fnlwgt']
    }).groupby('occupation')['fnlwgt'].sum() / cluster_data['fnlwgt'].sum()
    occ_weighted = occ_weighted.sort_values(ascending=False).head(3)
    
    results['occupation_unweighted'] = occ_unweighted
    results['occupation_weighted'] = occ_weighted
    
    return results

# MUESTRA 5K
print("MUESTRA 5K:")
print("-" * 70)

for cluster_id in sorted(sample_5k_data['cluster'].unique()):
    cluster_data = sample_5k_data[sample_5k_data['cluster'] == cluster_id]
    comparison = compare_weighted_unweighted(cluster_data)
    
    print(f"\nCLUSTER {cluster_id} (n={len(cluster_data)}):")
    print(f"  Peso poblacional total: {cluster_data['fnlwgt'].sum():,.0f}")
    print(f"  Porcentaje poblacional: {cluster_data['fnlwgt'].sum() / sample_5k_data['fnlwgt'].sum() * 100:.2f}%")
    
    # Ingresos
    print(f"\n  INGRESOS >50K:")
    print(f"    No ponderado: {comparison['income_unweighted']*100:.2f}%")
    print(f"    Ponderado:    {comparison['income_weighted']*100:.2f}%")
    print(f"    Diferencia:   {abs(comparison['income_weighted'] - comparison['income_unweighted'])*100:.2f} p.p.")
    
    # Sexo
    print(f"\n  DISTRIBUCIÓN POR SEXO:")
    print(f"    Hombres:")
    print(f"      No ponderado: {comparison['male_unweighted']:.2f}%")
    print(f"      Ponderado:    {comparison['male_weighted']:.2f}%")
    print(f"      Diferencia:   {abs(comparison['male_weighted'] - comparison['male_unweighted']):.2f} p.p.")
    print(f"    Mujeres:")
    print(f"      No ponderado: {comparison['female_unweighted']:.2f}%")
    print(f"      Ponderado:    {comparison['female_weighted']:.2f}%")
    print(f"      Diferencia:   {abs(comparison['female_weighted'] - comparison['female_unweighted']):.2f} p.p.")
    
    # Ocupación
    print(f"\n  OCUPACIÓN (Top 3):")
    for occ in comparison['occupation_unweighted'].index:
        pct_unweighted = comparison['occupation_unweighted'][occ] * 100
        pct_weighted = comparison['occupation_weighted'][occ] * 100
        print(f"    {occ.strip()}:")
        print(f"      No ponderado: {pct_unweighted:.2f}%")
        print(f"      Ponderado:    {pct_weighted:.2f}%")
        print(f"      Diferencia:   {abs(pct_weighted - pct_unweighted):.2f} p.p.")

print("\n" + "="*70)
print("MUESTRA 10K:")
print("-" * 70)

for cluster_id in sorted(sample_10k_data['cluster'].unique()):
    cluster_data = sample_10k_data[sample_10k_data['cluster'] == cluster_id]
    comparison = compare_weighted_unweighted(cluster_data)
    
    print(f"\nCLUSTER {cluster_id} (n={len(cluster_data)}):")
    print(f"  Peso poblacional total: {cluster_data['fnlwgt'].sum():,.0f}")
    print(f"  Porcentaje poblacional: {cluster_data['fnlwgt'].sum() / sample_10k_data['fnlwgt'].sum() * 100:.2f}%")
    
    # Ingresos
    print(f"\n  INGRESOS >50K:")
    print(f"    No ponderado: {comparison['income_unweighted']*100:.2f}%")
    print(f"    Ponderado:    {comparison['income_weighted']*100:.2f}%")
    print(f"    Diferencia:   {abs(comparison['income_weighted'] - comparison['income_unweighted'])*100:.2f} p.p.")
    
    # Sexo
    print(f"\n  DISTRIBUCIÓN POR SEXO:")
    print(f"    Hombres:")
    print(f"      No ponderado: {comparison['male_unweighted']:.2f}%")
    print(f"      Ponderado:    {comparison['male_weighted']:.2f}%")
    print(f"      Diferencia:   {abs(comparison['male_weighted'] - comparison['male_unweighted']):.2f} p.p.")
    print(f"    Mujeres:")
    print(f"      No ponderado: {comparison['female_unweighted']:.2f}%")
    print(f"      Ponderado:    {comparison['female_weighted']:.2f}%")
    print(f"      Diferencia:   {abs(comparison['female_weighted'] - comparison['female_unweighted']):.2f} p.p.")
    
    # Ocupación
    print(f"\n  OCUPACIÓN (Top 3):")
    for occ in comparison['occupation_unweighted'].index:
        pct_unweighted = comparison['occupation_unweighted'][occ] * 100
        pct_weighted = comparison['occupation_weighted'][occ] * 100
        print(f"    {occ.strip()}:")
        print(f"      No ponderado: {pct_unweighted:.2f}%")
        print(f"      Ponderado:    {pct_weighted:.2f}%")
        print(f"      Diferencia:   {abs(pct_weighted - pct_unweighted):.2f} p.p.")
```
